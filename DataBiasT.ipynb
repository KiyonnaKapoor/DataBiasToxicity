```python
comments_lowercase = test_df["Comment"]
comments_uppercase = []
i = 0 
while(i<33):
    comment = test_df["Comment"][i]
    comment=comment.upper()
    comments_uppercase.append(comment)
    i+=1

print(comments_uppercase)
```

```python
!pip install --upgrade google-api-python-client


from googleapiclient import discovery
import json


API_KEY = 'unreadable'
h=0 
toxicity_list = []
client = discovery.build( "commentanalyzer", "v1alpha1", developerKey=API_KEY, discoveryServiceUrl="https://commentanalyzer.googleapis.com/$discovery/rest?version=v1alpha1", static_discovery=False,)
while(h<33):

    analyze_request = {'comment': { 'text': comments_lowercase[h] },'requestedAttributes': {'TOXICITY': {}}}
    response = client.comments().analyze(body=analyze_request).execute()
    print(json.dumps(response, indent=2))
    toxicity_score = response['attributeScores']['TOXICITY']['summaryScore']['value']
    toxicity_list.append(toxicity_score)
    h+=1

```

```python
toxicity_list_uppercase = []
g=0
while(g<33):
    analyze_request_two = {'comment': { 'text': comments_uppercase[g] },'requestedAttributes': {'TOXICITY': {}}}
    response_two = client.comments().analyze(body=analyze_request_two).execute()
    print(json.dumps(response_two, indent=2))
    toxicity_score = response_two['attributeScores']['TOXICITY']['summaryScore']['value']
    toxicity_list_uppercase.append(toxicity_score)
    g+=1
print(toxicity_list_uppercase)
```

```python
mean_toxicity = sum(toxicity_list) / len(toxicity_list)
mean_toxicity_uppercase = sum(toxicity_list_uppercase) / len(toxicity_list_uppercase)
print(mean_toxicity)
print(mean_toxicity_uppercase)
```

```python
from matplotlib import pyplot as plt
plt.plot(toxicity_list)
plt.plot(toxicity_list_uppercase)
plt.xlabel('Comment Number')
plt.ylabel('Toxicity Score')
plt.title('Toxicity Scores of Comments in Lowercase and Uppercase')
```

```python
import numpy as np

difference_data = np.subtract(toxicity_list,toxicity_list_uppercase)
print(difference_data)
mean_toxicity3 = sum(difference_data) / len(difference_data)
```

```python
plt.plot(difference_data)
plt.xlabel('Comment Number')
plt.ylabel('Change in Toxicity Score')
plt.title('Change in Toxicity Scores of Comments in Lowercase and Uppercase')
```
